## 🎩 Mathematical Magic Tricks

**Mind-bending number patterns that seem like pure magic!**

</div>

````python
print("🎩 MATHEMATICAL MAGIC TRICKS:")
print("" + "="*45)

# 1089 Magic: Take any 3-digit number with different digits,
# reverse it, subtract smaller from larger, then reverse and add
# Always gives 1089!
magic_1089 = RPN("495 594 + 99 +")  # Example: 495 + 594 + 99 = 1089
result = magic_1089.eval()
print(f"🪄 1089 Magic trick result: {result} (always 1089!)")

# 6174 Kaprekar's constant: Take 4-digit number, arrange descending/ascending, subtract
# Eventually always reaches 6174
kaprekar_demo = RPN("8642 2468 -")  # Example step: 8642 - 2468 = 6174
result = kaprekar_demo.eval()
print(f"🔮 Kaprekar's constant demo: {result} (magic number 6174)")

# Multiplication by 9 finger trick: 9×7 = 63
# Hold up 10 fingers, fold down the 7th finger
# Left fingers = 6, right fingers = 3, answer = 63
nine_trick = RPN("9 7 *")
result = nine_trick.eval()
print(f"✋ 9×7 finger trick: {result} (6 fingers left, 3 right)")

print("\n🎪 Mathematics is pure magic! 🎪")
````

## 🔢 Amazing Number Patterns

**Patterns that make you go "HOW IS THIS POSSIBLE?!"**

</div>

````python
print("🔢 AMAZING NUMBER PATTERNS:")
print("" + "="*40)

# 37 × 3 = 111, 37 × 6 = 222, 37 × 9 = 333, etc.
print("🎯 The Magic of 37:")
pattern_37_3 = RPN("37 3 *")
pattern_37_6 = RPN("37 6 *")
pattern_37_9 = RPN("37 9 *")
print(f"   37 × 3 = {pattern_37_3.eval()}")
print(f"   37 × 6 = {pattern_37_6.eval()}")
print(f"   37 × 9 = {pattern_37_9.eval()}")

# 142857 × 2,3,4,5,6 creates cyclic permutations
print("\n🔄 Cyclic Number 142857:")
magic_142857_2 = RPN("142857 2 *")
magic_142857_3 = RPN("142857 3 *")
magic_142857_4 = RPN("142857 4 *")
print(f"   142857 × 2 = {magic_142857_2.eval()} (cyclic!)")
print(f"   142857 × 3 = {magic_142857_3.eval()}")
print(f"   142857 × 4 = {magic_142857_4.eval()}")

# Palindromic squares: 11² = 121, 111² = 12321
print("\n🪞 Palindromic Squares:")
palindrome_11 = RPN("11 2 **")
palindrome_111 = RPN("111 2 **")
palindrome_1111 = RPN("1111 2 **")
print(f"   11² = {palindrome_11.eval()} (palindrome!)")
print(f"   111² = {palindrome_111.eval()}")
print(f"   1111² = {palindrome_1111.eval()}")

print("\n🌟 Nature loves patterns! 🌟")
````

````python
print("🎲 CLASSIC CALCULATOR GAMES:")
print("" + "="*40)

# "Human Calculator" trick: Collatz conjecture demo
# Pick any number, if odd multiply by 3 and add 1, if even divide by 2
# Eventually reaches 1!
print("🧮 Collatz Conjecture Demo:")
collatz_demo = RPN("7 3 * 1 +")  # 7 → 22
step1 = collatz_demo.eval()
collatz_step2 = RPN("22 2 /")     # 22 → 11  
step2 = collatz_step2.eval()
print(f"   7 → {step1} → {step2} → ... → 1")
print("   (Try with any number - always reaches 1!)")

# Number guessing trick
print("\n🔮 Number Guessing Magic:")
print("   Pick 1-10, ×9, add digits, -5 = 4 (always!)")
print("   Then use 4 to pick predetermined answer")

# Birthday paradox calculator demo
print("\n🎂 Birthday Paradox:")
print("   23 people ≈ 50% chance of shared birthday")
print("   50 people ≈ 97% chance!")

print("\n🎪 Calculators: The original gaming device! 🎪")
````

## 🎪 Fun with Builder Pattern

**Modern RPN meets classic calculator fun!**

</div>

````python
print("🎪 FUN WITH BUILDER PATTERN:")
print("" + "="*42)

# Create a "magic" calculation that always gives interesting results
print("🪄 Magic Number Generator:")
magic_builder = (RPNBuilder()
    .var("x").push(9).mul()         # Multiply by 9
    .push(1).add()                  # Add 1 (simulating digit sum)
    .push(5).sub()                  # Subtract 5
    .abs())                         # Make positive

for x in [3, 7, 2, 9]:
    result = magic_builder.eval(x=x)
    print(f"   Magic with {x}: {result}")

# Digital root concept builder
print("\n🔮 Digital Root Concept:")
print("   Keep adding digits until single digit:")
print("   789 → 7+8+9=24 → 2+4=6")
print("   (Our RPN makes this calculation elegant!)")

# Fun calculator sequence
fun_sequence = (RPNBuilder()
    .push(12345)                    # Start with 12345
    .push(54321).add()              # Add its reverse
    .push(2).div()                  # Divide by 2
    .sqrt())                        # Square root

result = fun_sequence.eval()
print(f"\n🎯 Fun sequence: (12345+54321)/2 then √ = {result:.2f}")

print("\n🎊 Modern RPN keeps the calculator fun alive! 🎊")
````
